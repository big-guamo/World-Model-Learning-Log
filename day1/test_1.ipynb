# nn.Module 与优化器
掌握 nn.Linear、nn.Sequential、forward 写法，学习 torch.optim（SGD、Adam）和常见 loss 函数。把 Day 1 的线性回归改写为标准 nn.Module 风格

```python
import torch
import torch.nn as nn


# 1. 固定随机种子，方便复现
torch.manual_seed(0)


# 2. 构造数据
x = torch.linspace(-5, 5, 100).view(-1, 1)
y = 3 * x + 2 + torch.randn(x.shape) * 0.5


# 3. 定义模型
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()

        # 一元线性回归：输入 1 个特征，输出 1 个预测值
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        y_pred = self.linear(x)
        return y_pred


model = LinearRegressionModel()


# 4. 定义损失函数
criterion = nn.MSELoss()


# 5. 定义优化器
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)


# 6. 训练模型
epochs = 1000

for epoch in range(epochs):
    # 前向传播
    y_pred = model(x)

    # 计算损失
    loss = criterion(y_pred, y)

    # 清空上一轮梯度
    optimizer.zero_grad()

    # 反向传播，计算梯度
    loss.backward()

    # 更新参数
    optimizer.step()

    # 打印训练过程
    if (epoch + 1) % 100 == 0:
        w = model.linear.weight.item()
        b = model.linear.bias.item()

        print(
            f"Epoch [{epoch + 1}/{epochs}], "
            f"Loss: {loss.item():.4f}, "
            f"w: {w:.4f}, "
            f"b: {b:.4f}"
        )


# 7. 查看最终参数
print("训练完成")
print("学习到的 w:", model.linear.weight.item())
print("学习到的 b:", model.linear.bias.item())
```
